```python
import seaborn as sns
import pandas as pd
```

```python
data = pd.DataFrame({
    'A': [25, -10, 10, 5, 35, 13],
    'B': [0, 15, -5, 5, 20, 25],
    'C': [10, 25, -15, -5, -5, 15]
})
```

```python
sns.regplot(x="A", y="B", data=data)
sns.regplot(x="A", y="C", data=data)
sns.regplot(x="B", y="C", data=data)
```
